<h2>Scrape Wiki</h2>

```python
import requests
```

```python
wiki_link = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
raw_page = requests.get(wiki_link)
```

```python
page = raw_page.text
page = page.split('\n')
```

```python
import re
#header
for line in page :
    if "<th>" in line :
        print(line)
```

```python
table=[]
#body
for line in page :
    if "<td>" in line :
        line = re.sub('<.+?>','',line, 0).strip()
        table=table+[line]
table
```

```python
len(table)
```

```python

table2=[[0 for col in range(3)] for row in range(int(len(table)/3))]

i=0
for i in range(len(table)-1):
    table2[int(i/3)][i%3] = table[i]
    i = i+1

table2
```

```python
import pandas as pd
df=pd.DataFrame(table2)
df.columns =["Postcode", "Borough", "Neighbourhood"]
df
```

```python
# remove 'Borough'col is 'Not assigned'
df1= df[df.Borough!='Not assigned']
df1
```

```python
#remove same code
df2 = df1.drop_duplicates('Postcode')
df2
```
